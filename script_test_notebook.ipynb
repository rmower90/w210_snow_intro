```python
def get_all_and_baseline_pils(summary_table_fpath):
    # Load summary table
    df_summary_table = pd.read_csv(summary_table_fpath)
    
    # Convert time to datetime object
    df_summary_table['time'] = pd.to_datetime(df_summary_table['time'])
    
    # Create list of all pillows
    all_pils = df_summary_table.columns.to_list()
    all_pils.remove('time')
    all_pils.remove('aso_mean_bins_mm')
    
    # Identify baseline pillows
    df_year = df_summary_table.groupby(df_summary_table.time.dt.year)[all_pils].sum()
    pillow_w_flight_per_year = df_year.replace(0, pd.NA).dropna(axis=1,how='any').columns.to_list()
    pillows_cols = pillow_w_flight_per_year.copy()
    pillows_cols.append('time')
    slice_df = df_summary_table[pillows_cols]
    valid_time = slice_df.dropna(axis=0, how='any').time.values
    slice_df = df_summary_table[df_summary_table['time'].isin(valid_time)].dropna(axis=1, how='any')
    baseline_pils = slice_df.columns.to_list()
    baseline_pils.remove('time')
    baseline_pils.remove('aso_mean_bins_mm')
    
    return all_pils, baseline_pils
```

```python
summary_table_fpath = 'USCASJ/total.csv'
df = get_all_and_baseline_pils(summary_table_fpath)
all_pils = pd.DataFrame(df[0], columns=['id'])
baseline_pils = pd.DataFrame(df[1], columns=['id'])
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, CustomJS
from bokeh.io import output_notebook
import pandas as pd
# Sample data
data = {
    'x': pd.to_datetime(["2024-12-01", "2024-12-02", "2024-12-03", "2024-12-04"]),
    'snow_pillow': ['A', 'B', 'C', 'D'],
    'A': [1, 2, 3, 4],
    'B': [2, 3, 4, 5],
    'C': [3, 4, 5, 6],
    'D': [4, 5, 6, 7]
}
source = ColumnDataSource(data=data)
line_source = ColumnDataSource(data=dict(x=[], A=[], B=[], C=[], D=[]))
# Create a figure
p = figure(x_axis_type='datetime', title="Dynamic Line Chart", width=600, height=400)
# Add initial lines (empty)
lines = {}
for key in ['A', 'B', 'C', 'D']:
    lines[key] = p.line('x', key, source=line_source, line_width=2, legend_label=key)
# Callback to update the line source
callback_2 = CustomJS(args=dict(source=source, line_source=line_source, lines=lines), code="""
    const indices = cb_obj.indices;
    if (indices.length === 0) return;
    let selectedSnowPillows = [];
    let x_values = source.data['x'];
    let newData = { x: x_values };
    for (let i = 0; i < indices.length; i++) {
        const index = indices[i];
        const selectedSnowPillow = source.data['snow_pillow'][index];
        selectedSnowPillows.push(selectedSnowPillow);
    }
    for (let i = 0; i < selectedSnowPillows.length; i++) {
        const pillowKey = selectedSnowPillows[i];
        if (source.data[pillowKey]) {
            newData[pillowKey] = source.data[pillowKey];
        }
    }
    line_source.data = newData;
    line_source.change.emit();
""")
source.selected.js_on_change("indices", callback_2)
output_notebook()
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import Legend
# Sample data
x = [1, 2, 3, 4, 5]
y1 = []  # Line with zero data
y2 = []  # Line with actual data
# Create a new plot
p = figure(title="Bokeh Line Example", x_axis_label='x', y_axis_label='y')
# Add lines to the plot
line1 = p.line(x, y1, line_width=2, color="blue")
line2 = p.line(x, y2, line_width=2, color="green")
# Create a legend
legend_items = []
# Add legend items conditionally
if not all(v == 0 for v in y1):
    legend_items.append(("Line 1", [line1]))
    line1.visible = True
else:
    line1.visible = True
if not all(v == 0 for v in y2):
    legend_items.append(("Line 2", [line2]))
# Add the legend to the plot
legend = Legend(items=legend_items)
p.add_layout(legend, 'right')
# Show the plot
show(p)
```
